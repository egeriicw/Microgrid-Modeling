```python
# Library imports

from __future__ import annotations

from dataclasses import dataclass, asdict
from pathlib import Path
import hashlib
import json
import numpy as np
import pandas as pd
import pyarrow.dataset as ps
import glob
import os
import matplotlib.pyplot as plt
import random
import time
import math
import logging
import itertools

from openpyxl import load_workbook
import xlsxwriter as xw
```

```python
# Define the file pattern to search for the RESSTOCK *.Parquet files
DATA_DIR = "../data/"

# Define input data directories
INPUT_DATA_DIR = os.path.join(DATA_DIR, "input/")

# RESSTOCK directories
INPUT_DATA_DIR_RESSTOCK = os.path.join(INPUT_DATA_DIR, "resstock/")
INPUT_DATA_DIR_RESSTOCK_BUILDINGPROFILES = os.path.join(INPUT_DATA_DIR_RESSTOCK, "timeseries_individual_buildings/")

# Define output data directories

OUTPUT_DATA_DIR = os.path.join(DATA_DIR, "output/")
OUTPUT_DATA_DIR_SCENARIO_RUNS = os.path.join(OUTPUT_DATA_DIR, "scenario_runs/")
```

```python
weather_filename = 'LCD_USW00013743_2025.csv'

selected_columns =  ['DATE', 'HourlyDryBulbTemperature']

dc_temp = pd.read_csv(os.path.join(DATA_DIR + '/background/' + weather_filename), usecols=selected_columns, parse_dates=['DATE'])

dc_temp = dc_temp.set_index('DATE')
dc_temp['HourlyDryBulbTemperatureF'] = dc_temp['HourlyDryBulbTemperature'] * (9/5) + 32

dc_temp.head()

```

```python
dc_temp_resampled = dc_temp['HourlyDryBulbTemperatureF'].resample('h').mean()

dc_temp_resampled.head()

dc_temp_resampled.to_csv(os.path.join(DATA_DIR + '/background/' + 'dc_resampled_weather.csv'))

```
